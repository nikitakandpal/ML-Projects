Now we start with pre-processing


We have some skewed values like the total rooms, total bedrooms, population and household.

In stats we say its right skewed but not a guassian bell curve so what we do for those features is we are going to take the logrithm of those features and see what the distribution looks like by using that.


````python
train_data['total_rooms'] = np.log(train_data['total_rooms']+1)
train_data['total_bedrooms'] = np.log(train_data['total_bedrooms']+1)
train_data['population'] = np.log(train_data['population']+1)
train_data['households'] = np.log(train_data['households']+1)
````

````python
train_data.hist(figsize=(15,8))
````

normal distribution for the four features now.

Now we need to convert ocean proximity feature to numeric value to use it
Sometimes categorical features do not make sence to ordinally encode them so to not just give them numbers EG. just give 0, 1, 2, 3 values but to one hot and cold them. So to take the category vales and to turn them into binary features that can either be 0 or 1.

````python
train_data.ocean_proximity.value_counts()
````

We see the categories here.
Instead of taking and assigning numbers to them we are going to create a feature for every single one of them and we are going to give this a value of 1 or 0.
So we can say this is a feature it can either be yes/no meaning 1/0.

````python
pd.get_dummies(train_data.ocean_proximity)
````

Now we join it with train data and drop the ocean proximity as we have turned it into multiple features.

````python
train_data.join(pd.get_dummies(train_data.ocean_proximity)).drop(['ocean_proximity'],axis=1)
````

now save it in train_data.

````python
train_data = train_data.join(pd.get_dummies(train_data.ocean_proximity)).drop(['ocean_proximity'],axis=1)
````

Now lets look at feature engineering.

 Combine features to more interesting features.

 Eg. we have total rooms per block and also total bedrooms so we can combine them to have more interesting feature like number of bedrooms per room or how many rooms of those rooms are bedrooms.

 and we add one more feature that if we have a block with many households with more households than usual you will probably have more rooms so the rooms in and out themselves dont give us full picture.

````python
train_data['bedroom_ratio']= train_data['total_bedrooms']/train_data['total_rooms']
train_data['household_rooms']= train_data['total_rooms']/train_data['households']
````

````python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(),annot=True,cmap="YlGnBu")
````

We can add more features or sometimes we can remove remove features which have close to 0 correlation with the target variable

Now we train multiple models.

Regression model

````python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train,y_train = train_data.drop(['median_house_value'],axis=1),train_data['median_house_value']
X_train_s = scaler.fit_transform(X_train)

reg = LinearRegression()

reg.fit(X_train_s,y_train)
````

We need to split again into X and Y because we had split the data without the new features so now we do it again.

We dont have to do the train/test split again nut X/Y to add all these new features to the test data to evaluate the model

Also we are scaling the data to get better score.

X_train_s = scaler.fit_transform(X_train)

So now first we do is take the test data and do the same thing we did with training data

````python
test_data = X_train.join(y_train)

test_data['total_rooms'] = np.log(test_data['total_rooms']+1)
test_data['total_bedrooms'] = np.log(test_data['total_bedrooms']+1)
test_data['population'] = np.log(test_data['population']+1)
test_data['households'] = np.log(test_data['households']+1)

test_data = train_data.join(pd.get_dummies(test_data.ocean_proximity)).drop(['ocean_proximity'],axis=1)

test_data['bedroom_ratio']= test_data['total_bedrooms']/test_data['total_rooms']
test_data['household_rooms']= test_data['total_rooms']/test_data['households']
````

````python
X_test,y_test = test_data.drop(['median_house_value'],axis=1),test_data['median_house_value']
````

````python
X_test_s = scaler.fit_transform(X_test)
````

````python
reg.score(X_test_s,y_test)
````

Now we train another model Random Forest Model

````python
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor()

forest.fit(X_train_s,y_train)
````

````python
forest.score(X_test_s,y_test)
````

We can get even higher accuracy by providing a parameter grid and trying different options here. For that we are going to use grid search with cross validation. So cross validation to keep it simple, we take the data we split it into K folds K being a number and we use all but one fold to train the data and one for the evaluation. So if you have 10 folds we use 9 for training and 1 for testing and you do that for all possible combinations and then you evaluate the parameters that you used for this one run for this one iteration

Now,

We are importing grid search cross validation

And to this we need to provide a parameter grid and now here we specify the different hyperparameters that we want to try

First we tried with
param_grid = {

    "n_estimators": [3,10,30],

    "max_features": [2,4,6,8]

}

and got a worse result so now we try by adding more stuff to parameter grid

With this grid we can get the best optimal hyperparameter max_features=8, n_estimators=30 and we will check if this produces the best results. But it produces a result of 0.9526324779791201 which is worse so we add more parameters.

````python
from sklearn.model_selection import GridSearchCV

forest = RandomForestRegressor()

param_grid = {
    "n_estimators": [100,200,300],
    "min_samples_split":[2,4],
    "max_depth":[None,4,8]
}

grid_search = GridSearchCV(forest,param_grid, cv=5,
                           scoring="neg_mean_squared_error",
                           return_train_score=True)
grid_search.fit(X_train_s,y_train)
````

This is going to give us the optimal model.

We are going to get the best estimator the best regressor and then we can use this to score the best regressor against the test data, say when we want to launch this to production

It might sometimes even produce worse or it doesn't change much but more often it produces better results

````python
grid_search.best_estimator_
````

````python
grid_search.best_estimator_.score(X_test_s,y_test)
````
