<h3>We can using select_dypes() to seperate the data types into categorical and numeric datatypes for further analysis

```python
cat_cols=tiktok_data.select_dtypes(include=['object']).columns.tolist()
cat_cols
```

```python
num_cols=tiktok_data.select_dtypes(include=np.number).columns.tolist()
num_cols
```

<h3>Using Seaborn and Matplotlib libraries I tried to observe the correlation between the attributes using bivariate analysis from pairplot and heatmap to get the correlation values of each pair

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(13,17))
sns.pairplot(data=tiktok_data)
plt.show()
```

```python
plt.figure(figsize=(12, 7))
#sns.heatmap(tiktok_data.drop(['track_name','artist_name','album','time_signature','duration_ms'],axis=1).corr(), annot = True, vmin = -1, vmax = 1)
sns.heatmap(tiktok_data.corr(), annot = True, vmin = -1, vmax = 1,cmap = 'viridis')
plt.show()
```

<h3> Only numerical data are considered for distribution graphs ploting 15 graphs as 5*3 graphs for each attribute

```python

fig = plt.figure(figsize = (15, 15))
for i, a in enumerate(num_cols):
    ax = fig.add_subplot(5, 3, i+1)
    tiktok_data[a].hist(bins = 30, ax = ax, facecolor = 'green')
    ax.set_title(a, color = 'Black')
fig.tight_layout()  
plt.show()


```
